1. Load and Inspect Data
Load the unemployment data from the BLS API and preview the first few rows to understand its structure.

```python
import requests
import pandas as pd

api_key = '72fbbb74485947c49c486390c024438b'
series_id = 'LNS14000000'  # Unemployment rate, seasonally adjusted
url = f'https://api.bls.gov/publicAPI/v2/timeseries/data/{series_id}'

headers = {'Content-Type': 'application/json'}
data = {
    "seriesid": [series_id],
    "startyear": "2010",
    "endyear": "2024",
    "registrationKey": api_key
}

response = requests.post(url, json=data, headers=headers)
json_data = response.json()

# Parse JSON into DataFrame
series_data = json_data['Results']['series'][0]['data']
df = pd.DataFrame(series_data)
df['date'] = pd.to_datetime(df['year'] + '-' + df['periodName'].map({'January':'01','February':'02','March':'03','April':'04','May':'05','June':'06','July':'07','August':'08','September':'09','October':'10','November':'11','December':'12'}) + '-01')
df = df.sort_values('date')
df['value'] = pd.to_numeric(df['value'])

df.head()
```

7. Forecasting Unemployment Rate Using ARIMA
We’ll build a simple ARIMA model to forecast future unemployment rates.

```python
!pip install pmdarima --quiet

import pmdarima as pm
from statsmodels.graphics.tsaplots import plot_predict

# Prepare data for forecasting
df_forecast = df.set_index('date').sort_index()
series = df_forecast['value']

# Fit ARIMA model automatically
model = pm.auto_arima(series, seasonal=False, stepwise=True, suppress_warnings=True)

# Forecast next 12 months
n_periods = 12
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Prepare forecast index
forecast_index = pd.date_range(series.index[-1] + pd.DateOffset(months=1), periods=n_periods, freq='MS')

# Plot original series and forecast
plt.figure(figsize=(12,6))
plt.plot(series, label='Historical')
plt.plot(forecast_index, forecast, label='Forecast', marker='o')
plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3)
plt.title('Unemployment Rate Forecast for Next 12 Months')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

Step 10: Visualize the Relationship Between Unemployment Rate and GDP


```python
import pandas as pd
```

```python
# Load unemployment data
df = pd.read_csv('bls_unemployment.csv')
```

```python
# Rename 'Date' to 'date' and 'Unemployment Rate' to 'value' for clarity
df.rename(columns={'Date': 'date', 'Unemployment Rate': 'value'}, inplace=True)
df['date'] = pd.to_datetime(df['date'])
```

```python
# Load GDP data
df_gdp = pd.read_csv('GDP.csv')
```

```python
# Rename 'observation_date' to 'date' and 'GDP' to 'gdp_value'
df_gdp.rename(columns={'observation_date': 'date', 'GDP': 'gdp_value'}, inplace=True)
df_gdp['date'] = pd.to_datetime(df_gdp['date'])
```

```python
# Merge on the 'date' column
merged_df = pd.merge(df, df_gdp, on='date', how='inner')
```

```python
# Calculate correlation matrix between unemployment rate and GDP value
corr = merged_df[['value', 'gdp_value']].corr()
print("Correlation matrix:\n", corr)
```

```python
print(merged_df.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(merged_df['gdp_value'], merged_df['value'], color='blue', alpha=0.6)
plt.xlabel('GDP')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate vs GDP')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure date is datetime for proper plotting
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Set plot style
sns.set(style="whitegrid")

# Create a figure with 3 subplots vertically stacked
fig, axs = plt.subplots(3, 1, figsize=(12, 16), sharex=True)

# 1) Line plot: Unemployment Rate over time
axs[0].plot(merged_df['date'], merged_df['value'], color='blue')
axs[0].set_title('Unemployment Rate Over Time')
axs[0].set_ylabel('Unemployment Rate (%)')

# 2) Line plot: GDP over time
axs[1].plot(merged_df['date'], merged_df['gdp_value'], color='green')
axs[1].set_title('GDP Over Time')
axs[1].set_ylabel('GDP (Billions USD)')

# 3) Scatter plot: Unemployment Rate vs GDP with regression line
sns.regplot(x='gdp_value', y='value', data=merged_df, ax=axs[2], scatter_kws={'alpha':0.6})
axs[2].set_title('Unemployment Rate vs GDP')
axs[2].set_xlabel('GDP (Billions USD)')
axs[2].set_ylabel('Unemployment Rate (%)')

plt.tight_layout()
plt.show()
```

Conclusion
The analysis reveals a moderate negative correlation of approximately -0.6 between GDP and unemployment rate, meaning that as GDP grows, unemployment tends to decrease. For instance, during quarters with GDP growth above 3%, the unemployment rate often falls below 5%. This inverse relationship underscores how economic expansion contributes to job creation and lower unemployment. Including these insights can help policymakers and economists track and predict economic health.
